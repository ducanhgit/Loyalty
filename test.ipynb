```python
import requests
import pandas as pd
import random
```

```python
# Lấy sản phẩm
product_resp = requests.get('https://fakestoreapi.com/products')
products = product_resp.json()
```

```python
pd.DataFrame(products)
```

```python
products = product_resp.json()
product_price_map = {p['id']: p['price'] for p in products}

# Tạo đơn hàng giả cho 10 khách hàng
orders = []
for order_id in range(1, 21):  # 20 đơn
    user_id = random.randint(1, 10)
    products_in_order = []
    for _ in range(random.randint(1, 3)):
        product_id = random.choice(list(product_price_map.keys()))
        quantity = random.randint(1, 5)
        products_in_order.append({'productId': product_id, 'quantity': quantity})
    orders.append({'id': order_id, 'userId': user_id, 'products': products_in_order})
```

```python
pd.DataFrame(orders)
```
